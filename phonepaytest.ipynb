```python
#aggre_user

path1 = "C:/Users/aji09/OneDrive/Desktop/DS/Project/Phonepay/pulse/data/aggregated/user/country/india/state/"

agg_user_list= os.listdir(path1)

columns1={"States":[],"Years":[],"Quarter":[],"Brands":[],"Transaction_count":[],"Percentage":[]}

for states in agg_user_list:
    cur_states = os.path.join(path1, states)

    agg_year_list = os.listdir(cur_states)

    for year in agg_year_list:
        cur_year = os.path.join(cur_states, year)

        agg_file_list = os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)

            data = open(cur_file, "r")
            A = json.load(data)
            
            try: # why use try this file showing None
                for i in A ["data"]["usersByDevice"]:
                    brand=i["brand"]
                    count=i["count"]
                    percentage=i["percentage"]
                    columns1["Brands"].append(brand)
                    columns1["Transaction_count"].append(count)
                    columns1["Percentage"].append(percentage)
                    columns1["States"].append(states)
                    columns1["Years"].append(year)
                    columns1["Quarter"].append(int(file.strip(".json")))
            except:
                pass

agg_user = pd.DataFrame(columns1)

agg_user["States"]= agg_user["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar & Island')
agg_user["States"]= agg_user["States"].str.replace("-"," ")
agg_user["States"]= agg_user["States"].str.title()
agg_user["States"]= agg_user["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
# Table Creation

#sql connection
mydb = psycopg2.connect(host = "localhost",
                        user = "postgres",
                        port = 5432,
                        database = "Phonepay_data",
                        password = "your_password")
cursor = mydb.cursor()

#aggregated_user

create_query_1 = '''CREATE TABLE IF NOT EXISTS aggregated_user(States VARCHAR(250),
                                                                Years INT,
                                                                Quarter INT,
                                                                Brands VARCHAR(250),
                                                                Transaction_count BIGINT,
                                                                Percentage FLOAT)'''

cursor.execute(create_query_1)
mydb.commit()

insert_query_1 = '''INSERT INTO aggregated_user(States, Years, Quarter, Brands, Transaction_count, Percentage) VALUES(%s,%s,%s,%s,%s,%s)'''

data = agg_user.values.tolist()
cursor.executemany(insert_query_1, data)
mydb.commit()


#aggregated_transaction

mydb.rollback() # Reset failed transaction

create_query_2 = '''CREATE TABLE IF NOT EXISTS aggregated_transaction(States VARCHAR(250),
                                                                    Years INT,
                                                                    Quarter INT,
                                                                    Transaction_type VARCHAR(250),
                                                                    Transaction_count BIGINT,
                                                                    Transaction_amount BIGINT)'''

cursor.execute(create_query_2)
mydb.commit()

insert_query_2 = '''INSERT INTO aggregated_transaction(States, Years, Quarter, Transaction_type, Transaction_count, Transaction_amount) VALUES(%s,%s,%s,%s,%s,%s)'''

data = agg_trans.values.tolist()
cursor.executemany(insert_query_2, data)
mydb.commit()


#aggregated_insurance

create_query_3 = '''CREATE TABLE IF NOT EXISTS aggregated_insurance(States VARCHAR(250),
                                                                    Years INT,
                                                                    Quarter INT,
                                                                    Transaction_type VARCHAR(250),
                                                                    Transaction_count BIGINT,
                                                                    Transaction_amount BIGINT)'''

cursor.execute(create_query_3)
mydb.commit()

insert_query_3 = '''INSERT INTO aggregated_insurance(States, Years, Quarter, Transaction_type, Transaction_count, Transaction_amount) VALUES(%s,%s,%s,%s,%s,%s)'''

data = agg_insure.values.tolist()
cursor.executemany(insert_query_3, data)
mydb.commit()


#map_user

create_query_4 = '''CREATE TABLE IF NOT EXISTS map_user(States VARCHAR(250),
                                                        Years INT,
                                                        Quarter INT,
                                                        District VARCHAR(250),
                                                        RegisteredUsers BIGINT,
                                                        AppOpens BIGINT)'''

cursor.execute(create_query_4)
mydb.commit()

insert_query_4 = '''INSERT INTO map_user(States, Years, Quarter, District, RegisteredUsers, AppOpens) VALUES(%s,%s,%s,%s,%s,%s)'''

data = map_user.values.tolist()
cursor.executemany(insert_query_4, data)
mydb.commit()


#map_transaction

create_query_5 = '''CREATE TABLE IF NOT EXISTS map_transaction(States VARCHAR(250),
                                                                Years INT,
                                                                Quarter INT,
                                                                District VARCHAR(250),
                                                                Transaction_count BIGINT,
                                                                Transaction_amount BIGINT)'''

cursor.execute(create_query_5)
mydb.commit()

insert_query_5 = '''INSERT INTO map_transaction(States, Years, Quarter, District, Transaction_count, Transaction_amount) VALUES(%s,%s,%s,%s,%s,%s)'''

data = map_trans.values.tolist()
cursor.executemany(insert_query_5, data)
mydb.commit()


#map_insurance

create_query_6 = '''CREATE TABLE IF NOT EXISTS map_insurance(States VARCHAR(250),
                                                                Years INT,
                                                                Quarter INT,
                                                                District VARCHAR(250),
                                                                Transaction_count BIGINT,
                                                                Transaction_amount BIGINT)'''

cursor.execute(create_query_6)
mydb.commit()

insert_query_6 = '''INSERT INTO map_insurance(States, Years, Quarter, District, Transaction_count, Transaction_amount) VALUES(%s,%s,%s,%s,%s,%s)'''

data = map_insur.values.tolist()
cursor.executemany(insert_query_6, data)
mydb.commit()


#top_user

mydb.rollback() # Reset failed transaction

create_query_7 = '''CREATE TABLE IF NOT EXISTS top_user(States VARCHAR(250),
                                                        Years INT,
                                                        Quarter INT,
                                                        Pincodes INT,
                                                        RegisteredUsers BIGINT)'''

cursor.execute(create_query_7)
mydb.commit()

insert_query_7 = '''INSERT INTO top_user(States, Years, Quarter, Pincodes, RegisteredUsers) VALUES(%s,%s,%s,%s,%s)'''

data = top_user.values.tolist()
cursor.executemany(insert_query_7, data)
mydb.commit()


#top_transaction

mydb.rollback() # Reset failed transaction

create_query_8 = '''CREATE TABLE IF NOT EXISTS top_transaction(States VARCHAR(250),
                                                                Years INT,
                                                                Quarter INT,
                                                                Pincodes VARCHAR(6),
                                                                Transaction_count BIGINT,
                                                                Transaction_amount BIGINT)'''

cursor.execute(create_query_8)
mydb.commit()

top_trans["Pincodes"] = top_trans["Pincodes"].astype(str) # ensure pincode is string

insert_query_8 = '''INSERT INTO top_transaction(States, Years, Quarter, Pincodes, Transaction_count, Transaction_amount) VALUES(%s,%s,%s,%s,%s,%s)'''

data = top_trans.values.tolist()
cursor.executemany(insert_query_8, data)
mydb.commit()


#top_insurance

create_query_9 = '''CREATE TABLE IF NOT EXISTS top_insurance(States VARCHAR(250),
                                                                Years INT,
                                                                Quarter INT,
                                                                Pincodes VARCHAR(6),
                                                                Transaction_count BIGINT,
                                                                Transaction_amount BIGINT)'''

cursor.execute(create_query_9)
mydb.commit()

insert_query_9 = '''INSERT INTO top_insurance(States, Years, Quarter, Pincodes, Transaction_count, Transaction_amount) VALUES(%s,%s,%s,%s,%s,%s)'''

data = top_insu.values.tolist()
cursor.executemany(insert_query_9, data)
mydb.commit()
```
